# Train Test Split
You always want to evaluate your final model on a test set that hasnt been used at all in the training process. So we'll split off a test set here.


Note - Little less true if used cross-validation but is still considered a bad practice

````
from sklearn.model_selection import train_test_split
````

````
df_train,df_test = train_test_split(df,test_size = 0.25)
````

````
df_train.shape
````

````
df_train.head()
````

#Prepare our data for modelling
involves splitting the data back out into plain NumPy arrays.

````
X_train = df_train.drop(columns=["target", "target_name"]).values
Y_train = df_train["target"].values
````

````
Y_train
````

# Modeling - What is our baseline ?
simplest model we can think of?

if our baseline model is just randomly guessing the species of flower or guessing a single species for every data point we expect to have a model accuracy of 0.33 or 33% since we have different classes that are evenlt balanced.

so our models should atleast beat 33% accuracy

# Modelling -  Simple manual model
manualy look at our data and decide some cutoff points fro classification

````
def single_feature_prediction(petal_length):
    if petal_length < 2.5:
      return 0
    elif petal_length < 4.8:
      return 1
    else:
      return 2
````

````
X_train[:,2]
````

````
manual_y_predictions = np.array([single_feature_prediction(val) for val in X_train[:,2]])
````

````
manual_model_accuracy = np.mean(manual_y_predictions == Y_train)
````

````
print(f"Manual Model Accuracy: {manual_model_accuracy *100:.2f}%")
````

````
sns.scatterplot(x="petal length (cm)" , y = "petal width (cm)" , hue = "target_name" , data = df_predictions)
````

````
def plot_incorrect_predictions(df_predictions, x_axis_feature, y_axis_feature):
  fig,axs = plt.subplots(2,2,figsize=(10,10))
  axs = axs.flatten()
  sns.scatterplot(x=x_axis_feature , y=y_axis_feature, hue ="prediction_label" ,data=df_predictions , ax=axs[0])
  sns.scatterplot(x=x_axis_feature , y=y_axis_feature, hue ="target_name" ,data=df_predictions , ax=axs[1])
  sns.scatterplot(x=x_axis_feature , y=y_axis_feature, hue ="correct_predictions" ,data=df_predictions , ax=axs[2])
  axs[3].set_visible=False
  plt.show()
````

````
plot_incorrect_predictions(df_predictions, "petal length (cm)" , "petal width (cm)")
````
